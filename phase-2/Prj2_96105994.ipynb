```python
# setup preprocessor
preprocessor.set_params(
    vectorizer__n_components=256, # dimensionality of the word vectors
    vectorizer__w2v={
        'iter':128, # number of training iterations of the w2v model
        'min_count':2, # all words with total frequency lower than this will be ignored
    },
)

preprocessor.fit(train_data)
train_data = preprocessor.transform(train_data)
test_data = preprocessor.transform(test_data)
```

```python
# transform DataFrames of train set to numpy arrays for computing vectore functions mrore efficient.
X1_train = pd.DataFrame()
X1_train[[i for i in range(256)]] = pd.DataFrame(train_data['vec_1'].tolist(), index= train_data.index)
X1_train = pd.DataFrame(train_data['vec_1'].to_list(), columns=[i for i in range(256)])
x1_train = X1_train.to_numpy()
del X1_train

X2_train = pd.DataFrame()
X2_train[[i for i in range(256)]] = pd.DataFrame(train_data['vec_2'].tolist(), index= train_data.index)
X2_train = pd.DataFrame(train_data['vec_2'].to_list(), columns=[i for i in range(256)])
x2_train = X2_train.to_numpy()
del X2_train

Y_train = pd.DataFrame()
Y_train[['genre', 'popularity']] = train_data[['genre', 'popularity']]
y_train_genre = Y_train['genre'].to_numpy()
y_train_popularity = Y_train['popularity'].to_numpy()
del Y_train
```

```python
# prepared numpy arrays in train set
print('x1_train:\n', x1_train, x1_train.shape, '\n')
print('x2_train:\n', x2_train, x2_train.shape, '\n')
print('y_train_genre:\n', y_train_genre, y_train_genre.shape, '\n')
print('y_train_popularity:\n', y_train_popularity, y_train_popularity.shape, '\n')
```

```python
# transform DataFrames of test set to numpy arrays for computing vectore functions mrore efficient.
X1_test = pd.DataFrame()
X1_test[[i for i in range(256)]] = pd.DataFrame(test_data['vec_1'].tolist(), index= test_data.index)
X1_test = pd.DataFrame(test_data['vec_1'].to_list(), columns=[i for i in range(256)])
x1_test = X1_test.to_numpy()
del X1_test

X2_test = pd.DataFrame()
X2_test[[i for i in range(256)]] = pd.DataFrame(test_data['vec_2'].tolist(), index= test_data.index)
X2_test = pd.DataFrame(test_data['vec_2'].to_list(), columns=[i for i in range(256)])
x2_test = X2_test.to_numpy()
del X2_test

Y_test = pd.DataFrame()
Y_test[['genre', 'popularity']] = test_data[['genre', 'popularity']]
y_test_genre = Y_test['genre'].to_numpy()
y_test_popularity = Y_test['popularity'].to_numpy()
del Y_test
```

```python
# prepared numpy arrays in test set
print('x1_test:\n', x1_test, x1_test.shape, '\n')
print('x2_test:\n', x2_test, x2_test.shape, '\n')
print('y_test_genre:\n', y_test_genre, y_test_genre.shape, '\n')
print('y_test_popularity:\n', y_test_popularity, y_test_popularity.shape, '\n')
```

    برای مشاهده‌ی نمونه پیاده‌سازی این الگوریتم می‌توانید
    <a href="https://medium.com/@srishtisawla/introduction-to-naive-bayes-for-classification-baefefb43a2d">
    این مقاله
    </a>
را بررسی کنید.
    </font>

```python
import numpy as np
from scipy.stats import multivariate_normal, norm


class NaiveBayes:
    
    def __init__(self, kind='gaussian', smoothing=0):
        self.kind = kind
        self.smoothing = smoothing
        self.distribution = [] 
        
        
    def compute_params(self , x , y):
        if self.kind == 'gaussian':
            self.prior = np.array([x.shape[0] - np.sum(y), np.sum(y)]) / x.shape[0]
            self.mu = []
            self.cov = []

            self.mu.append(np.mean(x[y == 0], axis=0))
            self.mu.append(np.mean(x[y == 1], axis=0))
            self.cov.append(x[y == 0])
            self.cov.append(x[y == 1])
            self.cov[0] = np.cov(np.stack(self.cov[0], axis=1)) + self.smoothing * np.eye(self.cov[0].shape[1])
            self.cov[1] = np.cov(np.stack(self.cov[1], axis=1)) + self.smoothing * np.eye(self.cov[1].shape[1])
        
        if self.kind == 'bernoulli':
            self.prior = np.array([x.shape[0] - np.sum(y), np.sum(y)], dtype=np.float64) / x.shape[0]
            self.threshold = np.mean(x, axis=0)
            self.p = np.zeros((2, x.shape[1]))
    

    def fit(self, x, y):
        NaiveBayes.compute_params(self , x , y)
        classes = [0 , 1]
        if self.kind == 'gaussian':
            for label in classes :
                self.distribution.append(multivariate_normal(self.mu[label], self.cov[label]))

        if self.kind == "bernoulli":
            for label in classes:
                for j in range(x.shape[1]):
                    cnt = 0
                    to_check = x[y == label][:, j][x[y == label][:, j] > self.threshold[j]]
                    for i in to_check:
                        if i != 0:
                            cnt += 1
                    self.p[label][j] = (cnt + self.smoothing) /(x[y == label].shape[0] + 2 * self.smoothing)


    def predict(self, x):
        predictions = []
        for i in range(len(x)):
            p0 = np.log(self.prior[0])
            p1 = np.log(self.prior[1])
            
            if self.kind == "gaussian":
                p0 += self.distribution[0].logpdf(x[i]).sum()
                p1 += self.distribution[1].logpdf(x[i]).sum()
                
            if self.kind == "bernoulli":
                p0 += np.sum(np.log(self.p[0][x[i] > self.threshold])) + np.sum(np.log(1 - self.p[0][x[i] <= self.threshold]))
                p1 += np.sum(np.log(self.p[1][x[i] > self.threshold])) + np.sum(np.log(1 - self.p[1][x[i] <= self.threshold]))
           
            predictions.append(0 if p0 > p1 else 1)
               
        return np.array(predictions)
```

```python
import typing as th
from collections import Counter
from sklearn import metrics


class KNN:
    
    def __init__(self):
        pass

    def fit(self, x, y):
        self.x_train = x
        self.y_train = y

    def predict(self, x, k):
        self.k = k
        num_test = x.shape[0]
        num_train = self.x_train.shape[0]
        dists = np.zeros((num_test, num_train))
        
        threeSums = np.sum(np.square(x)[:,np.newaxis,:], axis=2) - 2 * x.dot(self.x_train.T) + np.sum(np.square(self.x_train), axis=1)
        dists = np.sqrt(threeSums)

        y_pred = np.zeros(num_test)
        for i in range(num_test):
            closest_y = []
            closest_y = [self.y_train[j] for j in np.argsort(dists[i])[0:self.k]]
            y_pred[i] = Counter(closest_y).most_common(1)[0][0]
        
        return y_pred
    
    def score(self, x, y, k):
        y_pred = self.predict(x, k)
        return metrics.accuracy_score(y, y_pred)
```

```python
#implementation of score functions. you just need to call the last function 'evaluate'. he will call his brothers =)))
import typing as th
from sklearn import metrics


def accuracy(y, y_hat) -> float:
    return metrics.accuracy_score(y, y_hat)

def f1(y, y_hat, alpha: float = 0.5, beta: float = 1.):
    return metrics.f1_score(y, y_hat, average=None, zero_division=0.)

def precision(y, y_hat) -> float:
    return metrics.precision_score(y, y_hat, average=None, zero_division=0.)

def recall(y, y_hat) -> float:
    return metrics.recall_score(y, y_hat, average=None, zero_division=0.)


evaluation_functions = dict(accuracy=accuracy, f1=f1, precision=precision, recall=recall)


def evaluate(y, y_hat) -> th.Dict[str, float]:
    """
    :param y: ground truth
    :param y_hat: model predictions
    :return: a dictionary containing evaluated scores for provided values
    """
    return {name: func(y, y_hat) for name, func in evaluation_functions.items()}
```

<h3> Clustering Score <\h3>

```python
import typing as th
from sklearn import metrics
from sklearn.metrics.cluster import adjusted_rand_score

def purity(y, y_hat) -> float:
    contingency_matrix = metrics.cluster.contingency_matrix(y, y_hat)
    return np.sum(np.amax(contingency_matrix, axis=0)) / np.sum(contingency_matrix)

def adjusted_rand_index(y, y_hat) -> float:
    return adjusted_rand_score(y, y_hat)

evaluation_functions = dict(purity=purity, adjusted_rand_index=adjusted_rand_index)

def evaluate(y, y_hat) -> th.Dict[str, float]:
    """
    :param y: ground truth
    :param y_hat: model predictions
    :return: a dictionary containing evaluated scores for provided values
    """
    return {name: func(y, y_hat) for name, func in evaluation_functions.items()}
```

    <h3>K-means
        (۱۵ نمره)
    </h3>
    الگوریتم K-means را در <code style="font-size:0.9em">src.clustering.models</code> از پایه پیاده‌سازی کرده و مقدار بهینه برای هایپرپرامتر‌های این خوشه‌بند (تعداد خوشه‌ها، تعداد گام‌ها و ...) را بر حسب معیار Purity  بدست‌ آورده و در نهایت بهترین تنظیمات و میزان عملکرد مربوط به آن را در کد زیر گزارش کنید.
     برای یادآوری شیوه‌ی عملکرد این الگوریتم می‌توانید
    <a href="https://medium.com/capital-one-tech/k-means-clustering-algorithm-for-machine-learning-d1d7dc5de882">
    این مقاله
    </a>
را بررسی کنید.
    <br>
    <b>توجه:</b>
    از آنجایی که برای بدست‌ آوردن بهترین پارامترها نیاز به محاسبه‌ی Purity دارید تابع مربوط به این معیار را در  <code style="font-size:0.9em">src.clustering.evaluation</code> تکمیل کنید.
    </font>
</div>

```python
from numpy import linalg as LA
from statistics import mean



class KMeans():
    
    def __init__(self, x, k):
        self.k = k
        self.x = x
        self.labels = np.random.randint(self.k, size=self.x.shape[0])
        self.centroids = self.x[np.random.choice(self.x.shape[0], self.k, replace=False)]
        

    def update_centroids(self):
        self.new_centroids = np.zeros(self.centroids.shape)
        cnt = [0 for j in range(self.k)]
        for i in range(len(self.x)):
            self.new_centroids[self.labels[i]] = self.new_centroids[self.labels[i]] + self.x[i]
            cnt[self.labels[i]] += 1
        for i in range(len(cnt)):
            self.new_centroids[i] /= cnt[i]
        return self.new_centroids
        
        
    def update_labels(self):
        for i in range(len(self.labels)):
            for j in range(len(self.centroids)):
                if LA.norm(self.x[i] - self.centroids[j]) < LA.norm(self.x[i] - self.centroids[self.labels[i]]):
                    if LA.norm(self.x[i] - self.centroids[j]) != 0:
                        self.labels[i] = j
                
                
    def fit(self, itr):
        for i in range(itr):
            self.update_labels()
            self.centroids = self.update_centroids()

    def predict(self):
        return self.labels
```

```python
kmn1 = KMeans(x1_cluster, 12)
kmn2 = KMeans(x2_cluster, 12)
kmn1.fit(50)
kmn2.fit(40)

print('first  vector:', evaluate(y_cluster_genre, kmn1.predict()))
print('second vector:', evaluate(y_cluster_genre, kmn2.predict()))

# todo: fit & validate the model to achieve the best purity on training data (vec_1 & vec_2)
# todo: report best settings
# target purity for vec_1>=0.23, and for vec_2>=0.30
```
